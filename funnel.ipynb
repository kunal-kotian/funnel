```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
```

```python
def UserSim(n, paramter):
    """
    input :
    n, the number to of users to simulate
    parameter, the parameter
    
    output: A list of exponential random variable simulations
    """
    return np.random.exponential(1/paramter,n) #1st parameter is scale = 1/parameter given aka lambda
```

1(a)



```python
n = 1000
parameter = 2
time_duration = UserSim(n,parameter)
```

```python
np.histogram(time_duration, bins=8)
```

```python
fig, ax = plt.subplots(1,1)
plt.hist(time_duration, rwidth=0.8,align='left')

plt.xlabel('Time')
plt.ylabel('Number of Users')
plt.title('Funnel Simulation')
plt.grid(False)

plt.show()
```
